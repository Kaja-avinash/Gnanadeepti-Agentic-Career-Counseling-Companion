# 🌟 Gnanadeepti 🧠 – Agentic Career Counseling Companion

*"Lighting the path to your dream career with clarity, courage, and code."*

An AI-powered intelligent assistant built on IBM Cloud to suggest personalized career paths based on academic stream, interest, and strengths.


```python
def career_advisor(stream, interest, strength, high_scores=[]):
    career_map = {
        "AI and DS": ["Data Scientist", "ML Engineer", "AI Researcher", "Data Analyst"],
        "AI and ML": ["Machine Learning Engineer", "AI Developer", "Research Scientist"],
        "CSM": ["Cybersecurity Analyst", "Network Architect", "System Administrator"],
        "CAI": ["Cognitive Computing Engineer", "NLP Specialist", "AI Support Engineer"],
        "CSE": ["Software Developer", "Full Stack Developer", "Cloud Engineer"],
        "IT": ["IT Support Specialist", "Systems Analyst", "Technical Consultant"],
        "Cybersecurity": ["Security Analyst", "Penetration Tester", "Risk Assessor"],
        "IoT": ["IoT Developer", "Embedded Systems Engineer", "IoT Architect"],
        "ECE": ["Electronics Engineer", "Communication Engineer", "VLSI Designer"],
        "EEE": ["Power Systems Engineer", "Control Systems Engineer", "Electrical Design Engineer"],
        "Mechanical": ["CAD Designer", "Automobile Engineer", "Production Engineer"],
        "Civil": ["Structural Engineer", "Site Engineer", "Urban Planner"],
        "Pharmacy": ["Pharmacovigilance Officer", "Clinical Research Associate", "Regulatory Affairs Specialist"]
    }

    interest_tags = {
        "problem-solving": ["AI Researcher", "Software Developer", "Cybersecurity Analyst", "Penetration Tester"],
        "coding": ["Full Stack Developer", "ML Engineer", "Embedded Systems Engineer", "Cloud Engineer"],
        "communication": ["Technical Consultant", "Data Analyst", "Site Engineer"],
        "research": ["AI Researcher", "NLP Specialist", "Research Scientist", "Clinical Research Associate"],
        "hardware": ["Electronics Engineer", "Control Systems Engineer", "VLSI Designer"],
        "networking": ["Network Architect", "IoT Architect", "Security Analyst"]
    }

    strength_tags = {
        "math": ["Data Scientist", "ML Engineer", "Power Systems Engineer", "AI Developer"],
        "logic": ["Cybersecurity Analyst", "Penetration Tester", "Software Developer"],
        "creativity": ["CAD Designer", "Full Stack Developer", "Urban Planner"],
        "analysis": ["Data Analyst", "Systems Analyst", "Pharmacovigilance Officer"],
        "communication": ["Technical Consultant", "Regulatory Affairs Specialist"]
    }

    suggested = set()

    if stream in career_map:
        suggested.update(career_map[stream])

    if interest in interest_tags:
        suggested.update(interest_tags[interest])

    if strength in strength_tags:
        suggested.update(strength_tags[strength])

    if high_scores:
        if "Maths" in high_scores:
            suggested.update(["Data Scientist", "AI Developer", "Control Systems Engineer"])
        if "Biology" in high_scores:
            suggested.update(["Clinical Research Associate", "Pharmacovigilance Officer"])
        if "English" in high_scores:
            suggested.update(["Technical Consultant", "Regulatory Affairs Specialist"])

    return sorted(suggested)
```

```python
stream = "AI and DS"
interest = "problem-solving"
strength = "coding"
high_scores = ["Maths", "English"]
```

```python
career_list = career_advisor(stream, interest, strength, high_scores)
```

```python
print("🎓 Career Recommendations:\n")
for i, job in enumerate(career_list, 1):
    print(f"{i}. {job}")
```
